```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,21,10)
y=x**3

plt.plot(x,y)
plt.title("Plot")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
fig = plt.figure(figsize=(10,2),dpi=100)
d1=fig.add_axes([1,1,1,1])
d1.plot(x,y)
d2=fig.add_axes([1.5,1.5,.1,.1])
d2.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.hist(y)
```

```python
plt.pie(x)
```
