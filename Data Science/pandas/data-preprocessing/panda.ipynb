```python
data=pd.read_csv("Nan.csv")
data.drop("Unnamed: 0",axis=1,inplace=True)
data.head()
```

```python
avg_loss=data['normalized-losses'].astype("float").mean()
data["normalized-losses"]=data["normalized-losses"].fillna(avg_loss)
```

```python
avg_stroke=data['stroke'].astype("float").mean()
data["stroke"]=data["stroke"].fillna(avg_stroke)
```

```python
avg_bore=data['bore'].mean()
data['bore']=data['bore'].fillna(avg_bore)

```

```python
avg_horsepower=data['horsepower'].mean()
data['horsepower']=data['horsepower'].fillna(avg_horsepower)
```

```python
data.head()
```

```python
data['num-of-doors'] = data['num-of-doors'].fillna(data["num-of-doors"].value_counts().idxmax())
```

```python
data["peak-rpm"]=data['peak-rpm'].fillna(data['peak-rpm'].value_counts().idxmax())
```

```python
data.dropna(axis=0,inplace=True)
```

```python
data.reset_index(drop=True,inplace=True)
```
