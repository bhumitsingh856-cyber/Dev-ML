```python
employees_data = {
    'Employee_ID': [101, 102, 103, 104, 105 ,106],
    'Name': ['Sophia', 'Liam', 'Mia', 'Noah', 'Emma',"Rod"],
    'Department': ['HR', 'Engineering', 'Marketing', 'Engineering', 'HR',"PR"],
    'Status': ['Full-Time', 'Full-Time', 'Contract', 'Full-Time', 'Contract',"CEO"]
}
sales_data = {
     'Employee_ID': [101, 102, 103, 104, 105],
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Revenue': [45000, 52000, 49000, 61000, 58000],
    'Expenses': [31000, 32500, 30000, 35000, 33000],
    'Online_Orders': [1200, 1450, 1300, 1900, 1750]
}
```

```python
df1=pd.DataFrame(employees_data)
df2=pd.DataFrame(sales_data)
df2
```

```python
pd.merge(df1,df2,how="inner")
```

```python
pd.merge(df1,df2,how="outer",on="Employee_ID")
```

```python
pd.merge(df1,df2,how="right",on="Employee_ID")
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df2.join(df1)
```

```python
data = {
    'Region': ['East', 'West', 'East', 'North', 'West', 'North', 'East', 'West', 'North', 'South'],
    'Manager': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob', 'Charlie', 'Alice', 'David', 'Charlie', 'David'],
    'Category': ['Electronics', 'Furniture', 'Clothing', 'Electronics', 'Clothing', 'Furniture', 'Furniture', 'Electronics', 'Clothing', 'Electronics'],
    'Sales': [12000, 22000, 8500, 15000, 9200, 18500, 11000, 26000, 7400, 19000],
    'Quantity': [15, 30, 45, 20, 25, 18, 12, 35, 40, 22]
}

df=pd.DataFrame(data)
df
```

```python
df.groupby("Region")["Sales"].sum()
```

```python
df.groupby(["Region","Category"])["Sales"].sum()
```

```python
df["Sales"].agg(["mean",'max','min'])
```

```python
pd.pivot_table(df,values="Sales",index="Region",columns="Category")
```

```python
pd.crosstab(df["Region"],df['Sales'])
```
