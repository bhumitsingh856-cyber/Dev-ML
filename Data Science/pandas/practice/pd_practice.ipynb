```python
import numpy as np
import pandas as pd 
```

### Feature Extraction

```python
data=pd.read_csv("anime.csv")
data.head()
```

```python
data.loc[0]['Title']
def extract(txt):
    a=''
    found=False
    for i in txt:
        if(i==")"):
            break
        if(i=="("):
            found=True
        if(found):
            a+=i
    return a[1:]

data['Episodes'] = data['Title'].apply(extract)
data['Episodes'] = data['Episodes'].str.replace(" eps","")
data['Episodes'] = data['Episodes'].astype(int)
data.head()
```

```python
def time(txt:str):
    idx=txt.index(")")
    return txt[idx+1:idx+20]
data['Timestamp']=data['Title'].apply(time)
data.head()
```
