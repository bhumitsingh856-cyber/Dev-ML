```python
import numpy as np
import pandas as pd 
```

```python
data=pd.read_csv("country.csv")
data.head()
```

```python
# country with highest population
data[data['population'] == data['population'].max()][['country', "population" , "capital_city"]]
```

```python
data[data['population'] == data['population'].min()][['country', "population" , "capital_city"]]
```

```python
# Top 5 countries with highest democracy_score
data.nlargest(5,"democracy_score")[['country',"democracy_score"]]
```

```python
# Total regions
for i,j in data.groupby("region"):
    print(i , end=" , ",)
data['region'].value_counts().count()
```

```python
# How many countries lie in Eastern Europe Region
data[data['region'] == "Eastern Europe"][['country']]
data['region'].value_counts()['Eastern Europe']
```

```python
# Leader of 2nd largest population country
data.nlargest(2,"population").iloc[-1][['country', "population" , "capital_city","political_leader","title"]]
```

```python
# Unknown leaders
data[data['political_leader'].isna()]['country']
```

```python
# Republic in name cpuntries
count = 0
def rc(txt):
    global count
    for i in txt:
        if("republic" in txt.lower()):
            count+=1
    return txt
data['country'].apply(rc)
count
     
```
