```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Distribution plot
- Distributed plots are for quantitative data
1. histplot/distplot
2. jointplot
3. pairplot
4. rugplot


```python
data=sns.load_dataset("tips")
data.head()
```

```python
sns.displot(data['total_bill'],bins=20,kde=True)
```

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="hex")
```

```python
sns.pairplot(data , hue="size",palette="rainbow")
```

```python
sns.rugplot(data)
```

# Categorical plot
- Categorical plots are for qualitative data
1. boxplot
2. violinplot
3. stripplot
4. swarmplot
5. barplot
6. countplot


```python
sns.boxplot(x="day",y='total_bill',data=data,palette="rainbow")
```

```python
sns.countplot(x=data['sex'],hue=data['smoker'])
```

```python
import numpy as np
sns.barplot(x=data['size'],y=data["total_bill"],estimator=np.sum)
```

```python
sns.violinplot(x="day",y='total_bill',data=data,palette="rainbow")
```

```python
sns.stripplot(x="day",y="total_bill",data=data)
```

```python
sns.swarmplot(x='size',y="total_bill",data=data,palette="rainbow")
```

```python
sns.violinplot(x='size',y="total_bill",data=data,palette="rainbow")
sns.swarmplot(x='size',y="total_bill",data=data,palette="rainbow")
```

```python
corr=data[['total_bill',"tip","size"]].corr()
sns.heatmap(corr,annot=True)
```

```python
sns.heatmap(data.pivot_table(values="total_bill",index="size",columns="sex"))
```

```python
sns.lmplot(x='total_bill',y='tip',data=data, palette='rainbow',hue="size")
```
